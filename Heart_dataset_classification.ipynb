```python
import pandas as pd
df=pd.read_csv("/content/Heart_attack.csv")
df



```

```python
df.isna().sum()
```

```python
df['BMI'].fillna(df['BMI'].median(),inplace=True)
```

```python
df['HeartDiseaseorAttack'].value_counts()
```

```python
x=df.drop('HeartDiseaseorAttack',axis=1)
y=df['HeartDiseaseorAttack']
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
obj1=StandardScaler()
x_train=obj1.fit_transform(x_train)
x_test=obj1.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model3=LogisticRegression()
lr_model3.fit(x_train,y_train)
predict1=lr_model3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(predict1,y_test)
print(accuracy*100)
```

```python
import pickle
with open('lr_model3.pkl','wb')as f:
  pickle.dump(lr_model3,f)
```
